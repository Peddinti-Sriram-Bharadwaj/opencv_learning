these are simple operations based on the image shape, normally performed on binary images. it needs two inputs, one is the input image and second is thje structuring element or the kernel which decide the nature of the operation. two basic morphological operators are erosion and dilation. 
Then, there are its variant forms like operning, closing , gradient etc. 

- Erosion. 

the basic idea of erosion is just like soil erosion. it erodes away the boundarus of foreground object ( a white foreground object). the kernel slides through the image as in a 2D convolution, and a pixel in the original image will be considered 1 only if all the pixels under the kernel is 1. otherwise it is eroded to 0. 

all the pixels near the boundary will be discarded depending upon the size of the kernel. so, the thickness of the foreground object decreses in the image. it is useful for removing some small white noises, and detach two connected objects etc/ 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img =cv2.imread('./sample_imgs/rasenshuriken.jpeg', 0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(erosion),plt.title('Eroded')
plt.xticks([]), plt.yticks([])
plt.show()
```

## dilation
it is the opposite of erosoin. here, a pixel elemetn is '1', if atleast one pixel under the kernel is '1'. so, it increases the white region in the image or size of rthe foreground image increases. normally, in cases like noise removal, erosion is followed by dilation. Because erosion removes white noises, but it also shrinks the object, so we dilate it. 

sine the noise is goine, they wont come back, nbunt our object area increases. it is also useful in joining broken parts of an object

```python
img =cv2.imread('./sample_imgs/rasenshuriken.jpeg', 0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dilation),plt.title('Dilation')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Opening

it is just the name of erosion followed by dilation

```python
img =cv2.imread('./sample_imgs/rasenshuriken.jpeg', 0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(opening),plt.title('Opening')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Closing
it is the reverse of Opening, Dilation followed by Erosion. 
it is useuful in closing small holes in the foreground objects or small black points in the object


```python
img =cv2.imread('./sample_imgs/rasenshuriken.jpeg', 0)
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(closing),plt.title('Closing')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Morphological gradient
it is the difference between dilaotin and erosion of an image. 



```python
img =cv2.imread('./sample_imgs/rasenshuriken.jpeg', 0)
kernel = np.ones((5,5),np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(gradient),plt.title('Gradient')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Top Hat. 
it is the difference between input image and the opening of the image. 

```python
img =cv2.imread('./sample_imgs/rasenshuriken.jpeg', 0)
kernel = np.ones((9,9),np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)


plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(tophat),plt.title('Tophat')
plt.xticks([]), plt.yticks([])
plt.show()
```

## black hat
it is the difference between closing of the input image and the input image. 


```python
img =cv2.imread('./sample_imgs/rasenshuriken.jpeg', 0)
kernel = np.ones((5,5),np.uint8)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)


plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(tophat),plt.title('Tophat')
plt.xticks([]), plt.yticks([])
plt.show()
```
