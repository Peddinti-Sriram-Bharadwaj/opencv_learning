## 2D convolution. 

As inone dimensional signals, images also can be filtered with various low-pass(LPF), high pass filters (HPF) etc. LPF helps in removing noises, blurring the images etc. HPF helps in finding edges in the images. 



OpenCV provides a function cv2.filter2D() to convolve a kenrel with an image. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./sample_imgs/rasenshuriken.jpeg')

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Image blurring (image smoothing)

it is achieved by convolving the image with. a low-pass filter kernel. it is useful for removing noises. it actually removes high frequncy content such as noise and edges from the image. So, edges are blurred out a little bit in this operation. there are other blurring techniques that do not blur the edges. 

there are four types of blurring techniques in OpenCV. 

- Averaging. 
this is done by convolving the image with a normalized box filter. It takes the average of all the pixels unde rthe kernel area and replae the central element. this is done by the function **cv2.blur()** **cv2.boxFilter**. 



```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./sample_imgs/rasenshuriken.jpeg')

blur = cv2.blur(img,(5,5))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

- Gaussian Blurring

in this, instead of box filter, gaussian kernel is used. It is done with trhe function cv2.GaussianBlur(). the height and width of the kernel is to be specified which should be positive and odd (odd to have a central peak). we should specify the standard deviation in X and Y direction, sigmaX and sigmaY respectively. Gaussian blurring is highly effective in removing gaussian noise from the image. 



```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./sample_imgs/rasenshuriken.jpeg')

blur = cv2.GaussianBlur(img,(5,5),0)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

- median blurring

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./sample_imgs/rasenshuriken.jpeg')

median = cv2.medianBlur(img,5)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

## bilateral filtering

it is a noise removal technique that keeps the edges sharp. whoever the operation is slower compared to the other filters. While the gaussian filter is a function of space alone, it doesnt consider whether pixels hae the ame intensity or not becuase of which the edges are blurred. 
bilateral filtering on the other hand will also take the pixel intensity difference into account, and only considers them for blurring. onoly those pixels with similar intensity to central pixel is considered for blurring. it preserves the edges since pixels at the edges have large intensity variation. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./sample_imgs/rasenshuriken.jpeg')

blur = cv2.bilateralFilter(img,9,75,75)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
