```python
import numpy as np
import cv2
```

## reading an image

cv2.imread() function reads an image. 

the second argument specifies the way image should be read. 

-1 :- loads a color image, any transparency will be neglected. it is the default flag. 

0 :- loads image into grayscale mode. 

1 :- loas image as it is, including the alpha channel 

```python
img = cv2.imread('./sample_imgs/rasenshuriken.jpeg',0)
```

## displaying an image

use the cv2.imshow() function to display an image in a window. the window automatically fits to the image size. 

firs argument is the window name passed as a string, and the second argument is an image. as many windows as required can be created but with different names. 



```python
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**cv2.waitKey()** is a keyboard binding function, the time is given in milliseconds. the function waits for specified milliseconds for any keyboard event. in that tiem, the progrma continues. it waits indefinitely for a key stroke if 0 is passed. It can also be set to detect specific key strokes. 

**cv2.destroyAllWindows** destroys all windoes we created. if  any specific window is to be destryoed, just pass the exact window name as an argument. 

### named window

sometimes, we can already create a window and load image to it later. in this case we have to specify if the window is resizable or not. By default it is autosize, if we give the flag :- *cv2.WINDOW_NORMAL*, we can resize the window. This will be helpful when image is too large in dimension and adding track bar to windows. 


```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Write an image

use the function **cv2.imwrite()** to save an image. 
the first argument ist he file name afnd the second arguemtn is the image you want to save ( second argument posses the variable name give in python code). 



```python
cv2.imwrite('rasenshuriken.png', img)
```

## wrapped up code 



```python
img = cv2.imread('./sample_imgs/rasenshuriken.jpeg',0)
cv2.imshow('image', img)

# use & 0xFF in 64 bit machines. no need in 32 bit. 
k = cv2.waitKey(0) & 0xFF
if k == 27:
    cv2.destroyAllWindows()
elif k == ord('s'):
    cv2.imwrite('rasenshuriken.png', img)
    cv2.destroyAllWindows()
```
