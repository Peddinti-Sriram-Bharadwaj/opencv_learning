## 2D histogram in OpenCV

for this, we need to convert the image from BGR to HSV

channels = [0,1]

bins = [180, 256]

range = [0,180,0,256]



```python
import cv2
import numpy as np

img = cv2.imread('../sample_imgs/rasenshuriken.jpeg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

hist = cv2.calcHist([hsv], [0,1], None, [180, 256], [0, 180, 0, 256])
```

## 2D histogram in Numpy



```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
img = cv2.imread('../sample_imgs/bright.jpeg')

# Convert to HSV color space
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Extract H and S channels
h, s, v = cv2.split(hsv)

# Compute 2D histogram
hist, xbins, ybins = np.histogram2d(h.ravel(), s.ravel(), bins=[180, 256], range=[[0, 180], [0, 256]])

# Plot the 2D histogram
plt.figure(figsize=(8, 6))
plt.imshow(hist.T, origin='lower', aspect='auto', cmap='inferno', extent=[0, 180, 0, 256], interpolation = 'nearest')
plt.colorbar(label='Frequency')
plt.xlabel('Hue')
plt.ylabel('Saturation')
plt.title('2D Histogram of Hue & Saturation')
plt.show()
```
