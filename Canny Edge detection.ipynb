It is a popular edge detection algorithm, devleoped by John F. Canny in 1986. It is a multistage algorithm. 

- stage 1 :- noise reduction 
- Stage 2 :- Finding intensity gradient of the image
- Stage 3:- non-maximum suppression
- stage 4:- hysteresis thresholding


## Noise reduction 
it is done becuase edge detection is susceptible to noise. the first step is to remove the noise in the image with a 5 x 5 gaussian filter. 

## finding the intensity gradient of the image. 
smoothened inage is then filtered with a sobel kernel in both directions to get the dervivate in the horizontal and vertical directions. from these two images, we can find the edge gradient and the vector for each pixel. 


gradient descent is always perpendicular to edges. it is rounded to one of the four angles representing vertical, horizontal and two diagonal directions. 

## non maximum suppression. 
after getting gradient magnitude and direction, a full scan of the image is done to remove any unwanted pixel which may not constitute the edge. for this, aqt every pixel it is checked if it is the local maximum in the direction of gradient. the result, we get a binary image with "thin edges". 

## Hysteresis Thresholding. 

this strage decides where all the values are really edges and which are not. for this, we need a minval and maxval threshold. any edges with intensity gradient more than maxval are sure to be edges, and those below minval fare sure to be discarded. those who lie in between are classified as edges or non-edges based on their connectivity. if they are connected to "sure-edges" pixels, they are considered to be part of edges. otherwise, they are also discarded. 

This stage also removes small pixel noises on the assumption that edges afre long lines. 




```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('./sample_imgs/rasenshuriken.jpeg',0)
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
