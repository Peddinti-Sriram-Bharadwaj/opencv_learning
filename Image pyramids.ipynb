sometimes, there is a necessity to work with images of different resolution of the smae image. for example, when searching for something in an image, like face, we are not sure at whiat size the object will be present in the image. in that case, we will need to create a set of images with different resolution and search for object in all images. these set of images of varying resolutions are called image pyramids. 

there are two kinds of image pyramids :- 
1) Gaussian Pyramid
2) Laplacidan Pyramid

Higher level in a gaussian pyramids is formed by removing consecutive rows and columns from a lower level image. Each pixel in higher level is formed by the contribution from 5 pixels in underlying level with gaussian weights. by doing so, a M x N image becomes a M/2 x N/2 image. 
 
So, the area reduces to one fourth of the original area. this is called an octave. the same pattern continues as we go upper in the pyramid. Similarly when expanding, area becomes 4 times in each level. we can find gaussian pyramids using cv2.pyrDown() and cv2.pyrUp() functions. 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./sample_imgs/rasenshuriken.jpeg')
lower_reso = cv2.pyrDown(img)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(lower_reso),plt.title('lower in pyramid')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
higher_reso = cv2.pyrUp(img)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(higher_reso),plt.title('higher in pyramid')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Laplacian pyramids
lapalcian pyramids are formed from gaussian pyramids, there is no exclusive function for that. lapalacian images are like edge images only, most of the elements are zero. 

they are used in image compression. A level in in laplacian pyramid is fomred by the difference between thaqt level in the gaussian pyrmaid and the expanded version of its upper level in gaussian pyramid. 

## Applications of pyramids

One major application of this is image blending. for ex:- in image stitching, the stacking of images may not look good due to discontinuities between images., IN that case, blending with images gives seamless blending without leaving much data in the images.
